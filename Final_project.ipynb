## Predictive modelling 

### Preprocessing

```python
# assign data to new dataset
airbnb_data_modelling = airbnb_data_eda
```

Columns id and host id will be removed. 

```python
airbnb_data_modelling = airbnb_data_modelling.drop(['id', 'host_id'], axis=1)
```

```python
airbnb_data_modelling.columns[airbnb_data_modelling.isnull().any()]
```

```python
def replace_missing_value(dataset, column):

        dataset[column].fillna(0, inplace=True)

replace_missing_value(airbnb_data_modelling, 'reviews_per_month')
```

```python
airbnb_data_modelling.columns[airbnb_data_modelling.isnull().any()]
```

Missing values in review per month has been replaced. Missing data for last review will be maintained for one hot encoding.  

### Feature engineering

```python
airbnb_data_modelling_fe = airbnb_data_modelling
```

Maximum value for minimum night and availability will be replaced with 365 as a full year time. 

```python
def replacing_year_values(dataset, column):

    dataset[column] = np.where(dataset[column] > 365, 365, dataset[column])
```

```python
replacing_year_values(airbnb_data_modelling_fe, 'minimum_nights')
replacing_year_values(airbnb_data_modelling_fe, 'availability_365')
```

Price range will be reduced according to 3 sigma rule 

```python
# calculate summary statistics
price_mean, price_std = mean(airbnb_data_modelling_fe['price']), std(airbnb_data_modelling_fe['price'])

# identify outliers
cut_off = price_std * 3
lower, upper = 0, price_mean + cut_off

print(f'Price_mean: {price_mean}')
print(f'Price_std : {price_std }')
print(f'Price bottom limit is ${lower } and top limit is ${upper }')

# identify outliers
outliers = [x for x in airbnb_data_modelling_fe['price'] if x > upper]
print(f'Identified outliers: {len(outliers)}')

# remove outliers
airbnb_data_modelling_fe.drop(airbnb_data_modelling_fe.loc[airbnb_data_modelling_fe['price'].isin(outliers)].index, inplace=True)

# replac 0 with 1 - rof ruthrer logatytmic tranformation 
airbnb_data_modelling_fe['price'] = np.where(airbnb_data_modelling_fe['price'] == 0, 1, airbnb_data_modelling_fe['price'])
```

```python
airbnb_data_modelling_fe['price'].hist(bins=10)
```

Price distribution skewed-right. To provide more symmetrical distribution logarithmic transformation will be applied

```python
airbnb_data_modelling_fe['price_log'] = np.log(airbnb_data_modelling_fe['price'])
airbnb_data_modelling_fe['price_log'].hist(bins=10)
```

For categorical columns (room type, neighbourhood) one hot encoding will apply.

```python
airbnb_data_modelling_ohe_room = airbnb_data_modelling_fe['room_type'].copy()

airbnb_data_modelling_ohe_room = pd.get_dummies(airbnb_data_modelling_fe['room_type'], prefix='room_type')
```

```python
# hyperparameter optimization function

def objective(space):  

    # optimized parameters
    xgb_params = {
        'max_depth': int(space['max_depth']),
        'colsample_bytree': space['colsample_bytree'],
        'learning_rate': space['learning_rate'],
        'subsample': space['subsample'],
        'random_state': int(space['random_state']),
        'min_child_weight': int(space['min_child_weight']),
        'reg_alpha': space['reg_alpha'],
        'reg_lambda': space['reg_lambda'],
        'n_estimators': 100,
        'objective': 'reg:squarederror'
    }
    
    model = xgb.XGBRegressor(**xgb_params)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # error for given parameters
    score = mean_squared_error(y_test, y_pred)
    
    print(f'Score: {score}')
    
    return{'loss':score, 'status': STATUS_OK , 'model':model}
    
# parameter space with different distributions (quniform,uniform,loguniform)   

space ={
    'max_depth': hp.quniform ('x_max_depth', 5, 20, 1),
    'colsample_bytree': hp.uniform ('x_colsample_bytree', 0.8, 1.),
    'learning_rate': hp.uniform ('x_learning_rate', 0.05, 0.2),
    'subsample': hp.uniform ('x_subsample', 0.7, 1.),
    'random_state': hp.quniform ('x_random_state', 0, 10000, 50),
    'min_child_weight': hp.quniform ('x_min_child_weight', 1, 10, 1),
    'reg_alpha': hp.loguniform ('x_reg_alpha', 0., 1.),
    'reg_lambda': hp.uniform ('x_reg_lambda', 0.7, 1.),
}

trials = Trials()

best_params = fmin(fn = objective, # optimization function defined above
            space=space, # parameter space
            algo=partial(tpe.suggest, n_startup_jobs=1),#algorithm used for optimization
            max_evals=10,# number of attempts
            trials=trials)

print("The best params: ", best_params)
```

```python
# function for printing learning curve 

def plot_learning_curve(model, X, y, cv, train_sizes):

    
    plt.figure(figsize=(12,8))
    plt.title(f'Learning curves for a {model}')

    
    train_sizes, train_scores, validation_scores = learning_curve(model, X ,  y , train_sizes = train_sizes,
    cv = cv, scoring = 'neg_mean_squared_error')
    
    train_scores_mean = -train_scores.mean(axis = 1)
    validation_scores_mean = -validation_scores.mean(axis = 1)
    
    plt.plot(train_sizes, train_scores_mean, label = 'Training error', color='r')
    plt.plot(train_sizes, validation_scores_mean, label = 'Validation error', color='g')
    plt.legend()
    
    return plt
```
